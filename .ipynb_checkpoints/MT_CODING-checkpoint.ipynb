```python
sigma_halfspace = 1e-2
fmax, fmin = 1e3, 1e-2
```

```python
def skin_depth(sigma, f):
    return 500./np.sqrt(sigma*f)
```

```python
skin_depth_min = skin_depth(sigma_halfspace, fmax)
skin_depth_max = skin_depth(sigma_halfspace, fmin)

print("The minimum skin depth is {:2.1f} m".format(skin_depth_min))
print("The maximum skin depth is {:2.1f} m".format(skin_depth_max))
```

```python
print(
    "The smallest cell size should be {:2.1f} m".format(
        skin_depth_min / 4.
    )
)
print(
    "The mesh should extend {:1.1e} m".format(
        skin_depth_max * 2.
    )
)
```

```python
cs = 39.  # core cell size
npad = 25  # number of padding cells
ncz = 100   # number of core cells

# define a tensor mesh
hz = [(cs, npad, -1.3), (cs, ncz)] 
M = discretize.TensorMesh([hz], x0='N')  # put the origin at the surface 

# plot the mesh
fig, ax = plt.subplots(1,1, figsize=(8, 3))
mesh.plotGrid(centers=True, faces=True, ax=ax)
ax.legend(["centers", "faces"])
ax.invert_xaxis()  # so that the surface is on our left hand side
ax.set_xlabel('z (m)')
ax.grid(which="both", linewidth=0.5)

print(
    "The mesh extends {:1.1e}m, is that far enough?".format(
        M.hx.sum()
    )
)
```

```python
print(M)
```

```python
sigma = np.ones(M.nC)*sigma_halfspace # conductivity values for all cells
mu = np.ones(M.nC)*mu_0 # magnetic permeability values for all cells
epsilon = np.ones(M.nC)*epsilon_0 # dielectric constant values for all cells

print(
    "There are {:1.0f} cell centers. \n"
    " sigma   is {:1.0f} elements long, all cells have a value of {:1.2e} S/m \n"
    " mu      is {:1.0f} elements long, all cells have a value of {:1.2e} H/m \n"
    " epsilon is {:1.0f} elements long, all cells have a value of {:1.2e} F/m \n".format(
        M.nC, 
        len(sigma), sigma_halfspace,
        len(mu), mu_0,
        len(epsilon), epsilon_0
    )
)
```

```python
frequency = 1e3 # Frequency (Hz)
omega = 2*np.pi*frequency # Angular frequency (rad/s)
```

```python
sigmahat = sigma  # quasi-static assumption
# sigmahat = sigma + 1j*epsilon*omega  # includes displacement current
```

```python
# Grad 
M.setCellGradBC([['dirichlet', 'dirichlet']]) # Setup boundary conditions
Grad = M.cellGrad # Gradient matrix

# MfMu
Mmu = utils.sdiag(mesh.aveCC2F * mu) 

# Mccsigma
Msighat = utils.sdiag(sigmahat) 

# Div
Div = M.faceDiv # Divergence matrix

# Right Hand Side
B = M.cellGradBC  # a matrix for boundary conditions
Exbc = np.r_[0., 1.] # boundary values for Ex
```

```python
# Assemble the matrix

# A-matrix
A = sp.vstack([
    sp.hstack([Grad, 1j*omega*Mmu]), # Top row of A matrix
    sp.hstack((Msighat, Div)) # Bottom row of A matrix
])

# Right-hand side
rhs = np.r_[
    -B*Exbc, 
    np.zeros(M.nC)
]
```

```python
%%time
Ainv = Solver(A) # Factorize A matrix
sol = Ainv*rhs   # Solve A^-1 rhs = sol
Ex = sol[:M.nC] # Extract Ex from solution vector u
Hy = sol[M.nC:M.nC+M.nN] # Extract Hy from solution vector u
```

```python
Zxy = - 1./Hy[-1] # Impedance at the surface

print("Impedance: {:1.1e} + {:1.1e}i".format(Zxy.real, Zxy.imag))
print("or in terms of Amplidude: {:1.1e} and phase: {:1.1f} degrees".format(
    np.absolute(Zxy), np.rad2deg(np.arctan(Zxy.imag / Zxy.real)))
)
```

```python
app_res = abs(Zxy)**2 / (mu_0*omega)
app_phase = np.rad2deg(np.arctan(Zxy.imag / Zxy.real))

print(
    "Apparent Resistivity: {:1.1f}, Phase: {:1.1f}".format(
        app_res, app_phase
    )
)
```

```python
print(
    "Note that the apparent resistivity, {:1.1f} "
    "is the same as the true half-space {:1.1f}".format(
        app_res,
        1./sigma_halfspace  
    )
)
```

```python
def simulation_mesh():

    dh = 5.0
    hx = [(dh, 5, -1.3), (dh, 100), (dh, 5, 1.3)]
    hy = [(dh, 5, -1.3), (dh, 100), (dh, 5, 1.3)]
    hz = [(dh, 5, -1.3), (dh, 100), (dh, 5, 1.3)]
    mesh = TensorMesh([hx, hy, hz], "CCC")

    return mesh
```

```python
mesh = simulation_mesh()
M = mesh

background_value = np.log(1.0 / 100.0)
block_value = np.log(1.0 / 50.0)

# Find cells below topography and define mapping

air_value = 0
ind_active = mesh.gridCC[:, 2] < 0.0
active_map = maps.InjectActiveCells(mesh, ind_active, air_value)

model = background_value * np.ones(ind_active.sum())

ind_block = (
    (mesh.gridCC[ind_active, 0] > -40.0) #Lebar
    & (mesh.gridCC[ind_active, 0] < 40.0)
    & (mesh.gridCC[ind_active, 1] > -30.0)
    & (mesh.gridCC[ind_active, 1] < 30.0)
    & (mesh.gridCC[ind_active, 2] > -200.0) #Tinggi
    & (mesh.gridCC[ind_active, 2] < -150.0)
)
model[ind_block] = block_value

# Define a single mapping from model to mesh
exponential_map = maps.ExpMap()
reciprocal_map = maps.ReciprocalMap()
model_map = active_map * reciprocal_map * exponential_map



# We can plot a slice of the model at Y=-2.5

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ind_slice = int(mesh.nCy / 2)
mesh.plotSlice(model_map * model, normal="Y", ax=ax, ind=ind_slice, grid=False)
ax.set_title("Model slice at y = {} m".format(mesh.vectorCCy[ind_slice]))
plt.show()
```

```python
# Setup the the survey object
# Receiver locations
rx_x, rx_y = np.meshgrid(np.arange(-600, 601, 100), np.arange(-600, 601, 100))
rx_loc = np.hstack(
    (utils.mkvc(rx_x, 2), utils.mkvc(rx_y, 2), np.zeros((np.prod(rx_x.shape), 1)))
 )

# Make a receiver list
rxList = []
for rx_orientation in ["xx", "xy", "yx", "yy"]:
    rxList.append(NSEM.Rx.Point3DImpedance(rx_loc, rx_orientation, "real"))
    rxList.append(NSEM.Rx.Point3DImpedance(rx_loc, rx_orientation, "imag"))
for rx_orientation in ["zx", "zy"]:
    rxList.append(NSEM.Rx.Point3DTipper(rx_loc, rx_orientation, "real"))
    rxList.append(NSEM.Rx.Point3DTipper(rx_loc, rx_orientation, "imag"))
```
